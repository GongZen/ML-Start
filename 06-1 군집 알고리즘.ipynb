```python
import urllib.request

# 파일 다운로드
url = 'http://bit.ly/fruits_300_data'
urllib.request.urlretrieve(url, 'fruits_300.npy')
print("다운로드 완료!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
fruits = np.load('fruits_300.npy')
```

```python
fruits.shape
```

```python
print(fruits[0, 0, :])
```

```python
plt.imshow(fruits[0], cmap='gray_r')
print(fruits[0, 50, :])
```

```python
fig, axs = plt.subplots(1, 2)
# 도화지 1장을 준비하고 그 안에 그림 그릴 칸을 1행 2열로 나누어라. subplot 안에 1과 2는 각각 행과 열의 개수를 의미한다. 
# 즉, 실행 결과로 전체 도화지 객체와 개별 칸들이 담긴 배열이라는 두 개의 덩어리를 묶어서 반환한다.
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
```

```python
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
apple
```

```python
print(apple.mean(axis=1))
```

```python
plt.hist(apple.mean(axis=1), alpha=0.5, label='apple')
plt.hist(pineapple.mean(axis=1), alpha=0.5, label='pineapple')
plt.hist(banana.mean(axis=1), alpha=0.5, label='banana')
plt.legend()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20,5))
axs[0].bar(range(10000), apple.mean(axis=0))
axs[1].bar(range(10000), pineapple.mean(axis=0))
axs[2].bar(range(10000), banana.mean(axis=0))
```

```python
fig, axs = plt.subplots(1, 2)
# 도화지 1장을 준비하고 그 안에 그림 그릴 칸을 1행 2열로 나누어라. subplot 안에 1과 2는 각각 행과 열의 개수를 의미한다. 
# 즉, 실행 결과로 전체 도화지 객체와 개별 칸들이 담긴 배열이라는 두 개의 덩어리를 묶어서 반환한다.
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
```

```python
apple_mean = apple.mean(axis=0).reshape(100, 100)
pineapple_mean = pineapple.mean(axis=0).reshape(100, 100)
banana_mean = banana.mean(axis=0).reshape(100, 100)
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].imshow(apple_mean, cmap='grey_r')
axs[1].imshow(pineapple_mean, cmap='grey_r')
axs[2].imshow(banana_mean, cmap='grey_r')
```

```python
abs_diff = np.abs(fruits - apple_mean)
#fruits와 apple_mean의 shape는 각각 어떠한가?
# (300, 100, 100)과 (100, 100)이 빼기(-)를 하면 어떻게 되는가?
abs_mean = np.mean(abs_diff, axis=(1,2))
print(abs_mean.shape)
```

```python
apple_index = np.argsort(abs_mean)[:100]
apple_index = apple_index.reshape(10, 10)
fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[apple_index[i, j]], cmap='grey_r')
        axs[i, j].axis('on')
plt.show()

```

```python
abs_diff = np.abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff, axis=(1, 2))
abs_mean
```

```python
banana_index = np.argsort(abs_mean)[:100]
banana_index = banana_index.reshape(10, 10)
fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[banana_index[i, j]], cmap='grey_r')
```
