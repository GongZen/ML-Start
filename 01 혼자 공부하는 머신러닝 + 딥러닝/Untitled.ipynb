```python
from pytrends import dailydata
import pandas as pd

# 1. 설정값 지정
keyword = '나 혼자만 레벨업'
# NaN 방지를 위해 2020년 12월부터 추출 (이월 데이터 역할)
start_year, start_mon = 2020, 12
stop_year, stop_mon = 2021, 12
# geo='' 설정으로 글로벌 전체 검색량 추출
geo_location = '' 

print(f"'{keyword}'의 글로벌 데이터를 추출 중입니다 (2020.12 ~ 2021.12)...")

try:
    # 2. 데이터 추출 (내부적으로 매달 데이터를 가져와 스케일링 수행)
    df = dailydata.get_daily_data(
        keyword, 
        start_year, start_mon, 
        stop_year, stop_mon, 
        geo=geo_location
    )

    if not df.empty:
        # 3. 데이터 슬라이싱 (2021년 1월 1일부터 12월 31일까지)
        # 2020년 12월은 '보정용'으로 사용된 후 여기서 제거됩니다.
        final_2021_df = df.loc['2021-01-01':'2021-12-31', [keyword]]
        
        # 4. CSV 파일 저장 (한글 깨짐 방지 인코딩)
        file_name = "Solo_Leveling_Global_SVI_2021_Clean.csv"
        final_2021_df.to_csv(file_name, encoding='utf-8-sig')
        
        print("-" * 30)
        print(f"성공! '{file_name}' 파일이 생성되었습니다.")
        print(f"결측치(NaN) 개수: {final_2021_df[keyword].isna().sum()}") # 0이어야 정상
        print(final_2021_df.head()) # 1월 1일부터 데이터가 꽉 차 있는지 확인
    else:
        print("데이터를 가져오는 데 실패했습니다.")

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
