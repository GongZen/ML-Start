```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3])
print('array1 type:', type(array1))
print('array1 array 형태:', array1.shape)

array2 = np.array([[1, 2, 3],
                   [2, 3, 4]])
print('array2 type:', type(array2))
print('array2 array 형태:', array2.shape)

array3 = np.array([[1, 2, 3]])
print('array3 type:', type(array3))
print('array3 array 형태:', array3.shape)
```

```python
# 여기서 중괄호 {}는 format() 함수의 인자가 치환되어 들어갈 자리를 지정하고 있다. 콜론(:)을 기준으로 왼쪽은 인덱스(순서), 오른쪽은 형식 지정자이다.
# ndim은 Numpy의 ndarray 객체가 가진 속성으로 해당 배열의 축(Axis)의 개수를 반환한다.
print('array1: {:0}차원, array2: {:1}차원, array3: {:2}차원'.format(array1.ndim, array2.ndim, array3.ndim))
```

```python
list1 = [1, 2, 3]
print(type(list1))
array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype)
```

```python
# int와 unsigned int의 차이는 무엇인가?
# int는 양수의 정수, 음수의 정수 모두 가능하나, unsigned int는 양수의 정수만 허용한다.
# int는 MSB(최상위 비트, 부호 비트)로 양수 또는 음수를 표현해야 하나, unsigned int는 양수만 가능하기 때문에 모든 비트를 숫자 크기로 표현할 수 있어, 양수 표현 범위가 2배가 된다.

list2 = [1, 2, 'test']
array2 = np.array(list2)
print(array2, array2.dtype)

list3 = [1, 2, 3.0]
array3 = np.array(list3)
print(array3, array3.dtype)
```

```python
array_int = np.array([1, 2, 3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)

array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1])
array_int2 = array_float1.astype('int32')
print(array_int2, array_int2.dtype)
```

```python
array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(-1, 3)
print(array2d)

array3 = array2d[[0, 1], 2]
print('array2d[[0, 1], 2] =', array3.tolist())

array4 = array2d[[0, 1], 0:2]
# ((0,0), (0,1)), ((1,0), (1,1))
print('array2d[[0, 1], 0:2] =', array4.tolist())

array5 = array2d[[0, 1]]
# ((0, :), (1, :))
print('array2d[[0, 1]] =', array5.tolist())

```

```python
# 불린 인덱싱의 장점은 [] 안에 array1d > 5 Boolean indexing을 바로 적용 가능하다는 것이다.
array3 = array1d[array1d>5]
print(array3)
```

```python
org_array = np.array([3, 1, 9, 5])
print('원본 행렬:', org_array)
# np.sort()로 정렬해보자.
sort_array1 = np.sort(org_array)
print('np.sort() 정렬 후:', sort_array1)
print('원본은:', org_array)
# ndarray.sort()로 정렬하면?
sort_array2 = org_array.sort()
print('반환된 행렬:', sort_array2)
print('그리고 원본은:', org_array)
```

```python
array2d = np.array([[8, 12],
                   [7, 1]])
sort_array2d_axis0 = np.sort(array2d, axis=0)
print('로우:\n', sort_array2d_axis0)

sort_array2d_axis1 = np.sort(array2d, axis=1)
print('칼럼:\n', sort_array2d_axis1)
```

```python
org_array = np.array([3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(sort_indices)
```

```python
name_array = np.array(['존', '마이크', '사라', '케이트', '사무엘'])
score_array = np.random.randint(70, 101, size=5)
print(score_array)

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 socre_array의 index:', sort_indices_asc)
print('ranking:', name_array[sort_indices_asc])

```
